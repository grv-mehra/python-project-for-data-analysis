# exploratory data analysis

```python
data.columns
```

```python
ax=sns.countplot(x='Gender',data=data)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
data.columns
```

```python
df=data.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values('Amount', ascending=False)
```

```python
ax=sns.barplot(x='Gender',y='Amount', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
data.columns
```

```python
sns.countplot(x='Age Group',data=data,hue='Gender')
```

```python
data.columns
```

```python
age_data=data.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values('Amount', ascending=False)
ax=sns.barplot(x='Age Group', y='Amount', data=age_data)
for bars in ax.containers:
    ax.bar_label(bars)
```

# top 10 states by orders

```python
data.columns
```

```python
s_vs_o=data.groupby(['State'],as_index=False)['Orders'].sum().sort_values(['Orders'], ascending=False).head(10)
sns.set(rc={'figure.figsize':(16,4)})
ax=sns.barplot(x='State', y='Orders', data=s_vs_o)
for bars in ax.containers:
    ax.bar_label(bars)
```

# total amount/sales from top 10 states

```python
s_vs_a=data.groupby(['State'],as_index=False)['Amount'].sum().sort_values(['Amount'], ascending=False).head(10)
sns.set(rc={'figure.figsize':(16,4)})
ax=sns.barplot(x='State', y='Amount', data=s_vs_a)
for bars in ax.containers:
    ax.bar_label(bars)
```

# product category

```python
sns.set(rc={'figure.figsize':(20,4)})
ax=sns.countplot(x='Product_Category', data=data)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sns.set(rc={'figure.figsize':(20,6)})
product_cat=data.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(['Amount'],ascending=False).head(10)
ax=sns.barplot(x='Product_Category',y='Amount', data=product_cat)
for bars in ax.containers:
    ax.bar_label(bars)
```
